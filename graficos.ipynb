# Graficos
Comparar os algoritmos em função no número do tempo de execução, da memória e da qualidade da solução.

* Para tal foram gerados 5 exemplos para cada tamanho da instância e então feita a média dos resultados.

## Branch And Bound

```python
import matplotlib.pyplot as plt
inf = 1800
x = [4,5,6,7,8,9,10] #tamanho das instacias

#Tempo
y_tempo_euc_bnb = [388.52,inf,inf,inf,inf,inf,inf]
y_tempo_man_bnb = [115.4,inf,inf,inf,inf,inf,inf]

plt.title('Gráfico do Tempo - BNB')
plt.xlabel('Instância (potencia de 2)')
plt.ylabel('Tempo (segundos)')

plt.plot(x, y_tempo_euc_bnb,color="red", label = "Euclidiana")
plt.plot(x, y_tempo_man_bnb,color="blue",label="Manhatan")

plt.rc('legend', fontsize=14, frameon=False)
plt.legend()
plt.show()

#Memoria
inf = 200510.80 * 10
y_mem_euc_bnb = [200510.80,inf,inf,inf,inf,inf,inf]
y_mem_man_bnb = [66783.80,inf,inf,inf,inf,inf,inf]

plt.title('Gráfico da Memória - BNB')
plt.xlabel('Instância (potencia de 2)')
plt.ylabel('Memória (quantidade nós)')

plt.yscale("log")
plt.plot(x, y_mem_euc_bnb,color="red", label = "Euclidiano")
plt.plot(x, y_mem_man_bnb,color="blue",label="Manhatan")

plt.rc('legend', fontsize=14, frameon=False)
plt.legend()
plt.show()


```

## Análise Comparativa

```python
#Todos os gráficos do estados explorados juntos

plt.title('Gráfico comparativo do Tempo - EUC')
plt.xlabel('Instância (potencia de 2)')
plt.ylabel('Tempo')

plt.plot(x, y_tempo_euc_bnb,color="green",label="bnb")
plt.plot(x, y_tempo_euc_tat,color="red",label="tat")
plt.plot(x, y_tempo_euc_chr,color="blue",label="chr")

plt.rc('legend', fontsize=14, frameon=False)
plt.legend()

plt.show()

plt.title('Gráfico comparativo do Tempo - MAN')
plt.xlabel('Instância (potencia de 2)')
plt.ylabel('Tempo')

plt.plot(x, y_tempo_man_bnb,color="green",label="bnb")
plt.plot(x, y_tempo_man_tat,color="red",label="tat")
plt.plot(x, y_tempo_man_chr,color="blue",label="chr")

plt.rc('legend', fontsize=14, frameon=False)
plt.legend()

plt.show()

plt.yscale("log")
plt.title('Gráfico comparativo da Memória - EUC')
plt.xlabel('Instância (potencia de 2)')
plt.ylabel('Memória')

plt.plot(x, y_mem_euc_tat,color="red",label="tat")
plt.plot(x, y_mem_euc_chr,color="blue",label="chr")

plt.rc('legend', fontsize=14, frameon=False)
plt.legend()

plt.show()

plt.yscale("log")
plt.title('Gráfico comparativo da Memória - MAN')
plt.xlabel('Instância (potencia de 2)')
plt.ylabel('Memória')

plt.plot(x, y_mem_man_tat,color="red",label="tat")
plt.plot(x, y_mem_man_chr,color="blue",label="chr")

plt.rc('legend', fontsize=14, frameon=False)
plt.legend()

plt.show()
```
